```python
#Preamble. Standard packages for to load
import astropy
from astropy.table import Table, Column, MaskedColumn
import numpy as np
from astroquery.vizier import Vizier
import matplotlib.pyplot as plt
import urllib2
# special IPython command to prepare the notebook for matplotlib
%matplotlib inline 
from astroquery.sdss import SDSS
from astropy import units as u
from astropy import coordinates as coords
from astropy.io import fits
import random
```

```python
catalog_list = Vizier.find_catalogs('New white dwarf SDSS DR12')
```

```python
print({k:v.description for k,v in catalog_list.items()})
```

```python
Vizier.ROW_LIMIT = -1
catalogs = Vizier.get_catalogs(catalog_list.keys())
```

```python
print(catalogs)
```

```python
catalogs[0]
```

```python
catalogs[0]['Coordinates'] = coords.SkyCoord(catalogs['J/MNRAS/455/3413/table6']['_RAJ2000'], catalogs['J/MNRAS/455/3413/table6']['_DEJ2000'], frame='icrs')
```

```python
PMF = catalogs[0]['PMF']
catalogs[0]['plate'] = MaskedColumn(length=len(catalogs[0]),dtype='int32')
catalogs[0]['mjd'] = MaskedColumn(length=len(catalogs[0]),dtype='int32')
catalogs[0]['fiber'] = MaskedColumn(length=len(catalogs[0]),dtype='int32')


for ind,obj in enumerate(PMF):
    split_PMF = obj.split('-')
    catalogs[0]['plate'][ind] = split_PMF[0]
    catalogs[0]['mjd'][ind] = split_PMF[1]
    catalogs[0]['fiber'][ind] = split_PMF[2]
```

```python
WD = catalogs[0]

WD
```

```python
directory = 'data/'
WD['file'] = MaskedColumn(length=len(catalogs[0]),dtype='S32')
for ind,plate in enumerate(catalogs[0]['plate']):
    spec = SDSS.get_spectra_async(plate=plate, mjd=WD['mjd'][ind], fiberID=WD['fiber'][ind])
    url_of_interest = str(spec[0]).split()[4]
    WD['file'][ind] = url_of_interest.split('/')[-1]
    get_file = urllib2.urlopen(url_of_interest)
    output_file = open(directory+url_of_interest.split('/')[-1], 'wb')
    output_file.write(get_file.read())
    output_file.close
```

```python
def plot_spec(num,directory,WD):
    fits_spec = fits.open(directory+WD['file'][num])
    wavelength = 10**fits_spec[1].data['loglam']
    flux = fits_spec[1].data['flux']
    fig, ax = plt.subplots(1, 2, figsize=(12,4))
    ax[0].plot(wavelength, flux)
    ax[1].plot(wavelength, flux)
    ax[1].set_xlim(3800, 4400)
    plt.show()
```

```python
plot_spec(0,directory,WD)
```

```python
indices = np.arange(len(good_WD))
np.random.shuffle(indices)
good_shuffle_WD = good_WD[indices]
validation_WD = good_shuffle_WD[0:299]
test_WD = good_shuffle_WD[300:599]
training_WD = good_shuffle_WD[600:]
```

```python
plt.errorbar(training_WD['logg'], np.log(training_WD['Teff']), xerr=training_WD['e_logg'], 
             yerr=training_WD['e_Teff']/training_WD['Teff'], fmt=None, capsize=0, label='train')
plt.errorbar(test_WD['logg'], np.log(test_WD['Teff']), xerr=test_WD['e_logg'], 
             yerr=test_WD['e_Teff']/test_WD['Teff'], fmt=None, capsize=0, label='test')
plt.errorbar(validation_WD['logg'], np.log(validation_WD['Teff']), xerr=validation_WD['e_logg'], 
             yerr=validation_WD['e_Teff']/validation_WD['Teff'], fmt=None, capsize=0, label='val')
plt.legend()


plt.ylabel('Log(T)')
plt.xlabel('Log(g)')
```
